```python
#@title Editing synthetic images with CSG

prompt = "a picture of a cat"#@param {"type": "string"}
task_name = "cat2dog" #@param {"type": "string"}
save_dir = "output/synth_edit" #@param {"type": "string"}
guidance_scale = 0.0 #@param {"type": "number"}
random_seed = 0 #@param {"type": "integer"}
display_result = True #@param {"type": "boolean"}

test_cmd = f'''python src/edit_synthetic_images.py \
    --results_folder "{save_dir}" \
    --prompt_str "{prompt}" \
    --task "{task_name}" \
    --random_seed {random_seed} \
    --mask_res 16 --posterior_guidance {guidance_scale}'''

with open('./synth_edit_script.sh', 'w') as file:
    file.write(test_cmd)

!bash synth_edit_script.sh

if display_result:
    from IPython.display import Image
    import os
    recon = Image(os.path.join(save_dir, "reconstruction.png"))
    display(recon)
    edit = Image(os.path.join(save_dir, "edit.png"))
    display(edit)
    mask = Image(os.path.join(save_dir, "mask.png"))
    display(mask)
```
